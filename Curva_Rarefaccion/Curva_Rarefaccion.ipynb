```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tb_ddbb_file = "SNP_monitoring_curve_20250402_225811_without_complex.tsv"
snps_tb_ddbb = pd.read_csv(tb_ddbb_file, sep='\t')

plt.figure(figsize=(10,6))
plt.plot(snps_tb_ddbb["PACK"], snps_tb_ddbb["num_SNPs"], color='black', label="SNPs totales = 2.201.905")

x = np.linspace(0, snps_tb_ddbb["PACK"].max(), 100)
slope = snps_tb_ddbb["num_SNPs"].max() / snps_tb_ddbb["PACK"].max()
y = slope * x
plt.plot(x, y, color='red', ls='--', lw=1, label='Lnea de referencia')

plt.axvspan(0, 222, facecolor='#88CCEE', alpha=0.25)

# Leyenda y rtulos
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.xlabel("Nmero de paquetes de 500 muestras")
plt.ylabel("SNPs totales aportados")
#plt.title("Curva de rarefaccin")

plt.grid()
plt.show()
```

```python
print("Mximo eje X (PACK):", snps_tb_ddbb["PACK"].max())
print("Mximo eje Y (num_SNPs):", snps_tb_ddbb["num_SNPs"].max())
```
